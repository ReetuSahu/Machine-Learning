```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("C:/Users/lkosh/Desktop/MachineLearning/Cases/santander-customer-satisfaction/train.csv")
```

```python
train = train.drop(['ID'],axis=1)
```

```python
X = train.drop(["TARGET"],axis=1)
y = train['TARGET']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
train_scaled=scaler.transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principalComponents = pca.fit_transform(train_scaled)
```

```python
print(pca.explained_variance_)
```

```python
print(np.sum(pca.explained_variance_))
```

```python
print(pca.explained_variance_ratio_) 
```

```python
print(pca.explained_variance_ratio_ * 100) 
```

```python
import matplotlib.pyplot as plt
ys = np.cumsum(pca.explained_variance_ratio_ * 100)
xs = np.arange(1,370)
plt.plot(xs,ys)
plt.show()

```

```python
ys
```

```python
np.sum(ys<=91)
```

```python
cols = []
for i in range(1,76):
    name = "PC" + str(i)
    cols.append(name)
```

```python
# principalComponents are PCA scores

X_PCS = pd.DataFrame(principalComponents[:,:75],
                 columns = cols)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(X_PCS,y)
```

```python
test = pd.read_csv("C:/Users/lkosh/Desktop/MachineLearning/Cases/santander-customer-satisfaction/test.csv")
```

```python
test_X = test.iloc[:,1:]
```

```python
test_X_scaled = scaler.transform(test_X)
```

```python
principle_component_test = pca.transform(test_X_scaled)
```

```python
X_test_PCS = pd.DataFrame(principle_component_test[:,:75],columns = cols)
```

```python
y_pred_prob = logreg.predict_proba(X_test_PCS)[:,1]
```

```python
source = pd.DataFrame({"ID":test["ID"],"TARGET":y_pred_prob})
source.to_csv("C:/Users/lkosh/Desktop/MachineLearning/Cases/santander-customer-satisfaction/Score_PCA.csv",index=False)
```
